**Tensorflow NN with sequential and functional API**

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python
#hardware_devices = tf.config.list_physical_devices('GPU')
#ardware_devices[0]
```

```python
#con = tf.config.experimental.set_memory_growth(hardware_devices[0], True)
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(type(x_train))
```

**Let's reshape the above array, for now we need an array of 28*28 as input to our sequential model input layer**

```python
x_train = x_train.reshape(-1, 28*28).astype('float32')/255.0
x_test = x_test.reshape(-1, 28*28).astype('float32')/255.0
```

```python
x_train.shape
```

**Using Sequential API of Keras for creating NN**

**without hidden layers**

```python

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=3)
```

image.png

```python
model.evaluate(x_test, y_test)
```

```python
y_predicted = model.predict(x_test)
y_predicted[0]
```

```python
import numpy as np
np.argmax(y_predicted[0])
```

```python
x_test[0]
```

**with Hidden layer**

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test, y_test)
```

**we can also replace input_shape argument with flatten layer**

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(784,)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test, y_test)
```

**Let's use functional API**

```python
input_layer = keras.Input(shape=(784,))
first_layer = layers.Dense(256, activation = 'relu', name="layer_1")(input_layer)
second_layer = layers.Dense(128, activation = 'relu', name = 'layer_2')(first_layer)
final_layer = layers.Dense(10, activation='softmax', name = 'output_layer')(second_layer)
model = keras.Model(inputs=input_layer, outputs=final_layer)
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test, y_test)
```
